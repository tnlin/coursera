```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr = np.loadtxt("data1.txt", delimiter=',')
x = arr[:,0]
y = arr[:,1]
# Append ones for gradient descent
X = np.c_[ np.ones(x.shape[0]), x ]
```

```python
%matplotlib inline
plt.ylabel("Profit in $10,000s")
plt.xlabel("Population of City in 10,00s")
plt.legend()
plt.grid()
plt.plot(x, y, 'rx')
```

```python
from gradientDescent import gradientDescent
```

```python
theta = [0] * 2
alpha = 0.01
num_iters = 1000
theta, J_history = gradientDescent(X, y, theta, alpha, num_iters)
print(theta)
```

```python
%matplotlib inline
plt.ylabel("Profit in $10,000s")
plt.xlabel("Population of City in 10,00s")
plt.legend()
plt.grid()
plt.plot(x, y, 'rx')
plt.plot(x, np.dot(X, theta))
```

```python
%matplotlib inline
plt.ylabel("Cost")
plt.xlabel("Iters")
plt.legend()
plt.grid()
plt.plot(J_history)
```
